```python
bk = rental[rental['Boro']=='Brooklyn']
```

```python
bk = bk[(bk['Neighborhood']!='Murray Hill') & (bk['Neighborhood']!='East Village')]
```

```python
hood_df = bk.groupby(['Agency','Neighborhood','Beds']).agg({'Price':'mean'}).sort_values(by='Price',ascending=False)
```

```python
hood_df.reset_index(inplace=True)
```

```python
hood_df['Num_beds'] = hood_df['Beds'].apply(lambda x: x.split()[0])
```

```python
hood_df['Num_beds'] = hood_df.Num_beds.replace('Studio','1')
```

```python
hood_df['Num_beds'] = pd.to_numeric(hood_df['Num_beds'])
```

```python
hood_df['Price_Bed'] = hood_df['Price']/hood_df['Num_beds']
```
